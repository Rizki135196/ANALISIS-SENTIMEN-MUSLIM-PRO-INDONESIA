```python
from google_play_scraper import Sort, reviews_all
import csv
import pandas as pd
```

```python
# Mengambil ulasan dari aplikasi Muslim Pro Indonesia
from google_play_scraper import reviews_all, Sort

scrapreview = reviews_all(
    'com.bitsmedia.android.muslimpro',  # ID aplikasi Muslim Pro Indonesia
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=150000
)

scrapreview_limited = scrapreview[:150000]
```

```python
# 1. Hasil scraping
scrapreview = [
    {'content': 'Bagus sekali', 'score': 5},
    {'content': 'Biasa aja', 'score': 3},
    {'content': 'Aplikasi buruk', 'score': 1}
]

# 2. Batasan review
scrapreview_limited = scrapreview[:15000]

# 3. Buat DataFrame
import pandas as pd
dataView = pd.DataFrame(scrapreview_limited)
dataView.head()
```

```python
import csv
with open('hasil_scraping_muslimpro.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['content', 'score'])

    for review in scrapreview_limited:
        writer.writerow([
            review['content'],
            review.get('score', '')
        ])

print("Data telah disimpan ke hasil_scraping_muslimpro.csv")
```
